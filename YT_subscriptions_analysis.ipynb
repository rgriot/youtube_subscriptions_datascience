```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path_to_data = "./channels_data"
sub_channels = "subscribed_channels.csv"
all_files = os.listdir(path_to_data)
```

```python
def import_text(files, text = "video_title"):
    channel_text = {}
    for file in all_files:
        current_data = pd.read_csv(f"{path_to_data}/{file}")
        info = list(current_data[text])
        current_channel = current_data["channel_id"][0]
        channel_text[current_channel] = " ".join(info)
    return channel_text

def channel_id_to_title(video_dict):
    channels = pd.read_csv(f"./{sub_channels}")
    new_dict = {}
    
    for channel_id in video_dict.keys():
        channel_title = channels.loc[channels["channel_id"] == channel_id, "channel_title"].values[0]
        new_dict[channel_title] = video_dict[channel_id]
    return new_dict
```

```python
video_titles = import_text(all_files)

video_titles = channel_id_to_title(video_titles)
```

```python
channels = pd.read_csv(f"./{sub_channels}")

channels
```

```python
channel_theme = [
    "ecology",
    "ecology",
    "video_games",
    "rpg",
    "spearfishing",
    "cooking",
    "survival",
    "survival",
    "fitness",
    "pop_culture",
    "didgeridoo",
    "spearfishing",
    "didgeridoo",
    "minecraft",
    "survival",
    "news",
    "starcraft",
    "minecraft",
    "minecraft",
    "news",
    "economy",
    "science",
    "fitness",
    "survival",
    "minecraft",
    "video_games",
    "science",
    "pop_culture",
    "spearfishing",
    "cinema",
    "minecraft",
    "comedy",
    "survival",
    "cooking",
    "grooming",
    "video_games",
    "personal_dev",
    "ecology",
    "fitness",
    "ecology"
]

channels["channel_theme"] = channel_theme

correspondance_dict_df = pd.DataFrame()
correspondance_dict_df["channel_title"] = video_titles.keys()
correspondance_dict_df = pd.merge(correspondance_dict_df, channels, on = "channel_title", how = "left")
correspondance_dict_df = correspondance_dict_df[["channel_title", "channel_theme"]]
```

```python
import nltk
import string
from nltk.stem.snowball import FrenchStemmer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.manifold import TSNE

stemmer = FrenchStemmer()
tokenizer = nltk.RegexpTokenizer(r"\w+")


def get_tokens(video_text):
    tokens = dict()
    
    for video, desc in video_text.items():
        tokens[video] =  desc.lower().translate(str.maketrans("", "", string.punctuation))
    return tokens

def get_stem_tokens(tokens, stemmer):
    sw = set()
    sw.update(tuple(nltk.corpus.stopwords.words('french')))
    sw.update(tuple(nltk.corpus.stopwords.words('english')))
    
    stemmed = [stemmer.stem(w) for w in tokens if w not in list(sw)]

    return stemmed

def tokenize(text):
    tokens = nltk.word_tokenize(text)
    stems = get_stem_tokens(tokens, stemmer)
    return stems

def label_point(x, y, val, ax, dx):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'] + dx, point['y'], str(point['val']))


def TSNE_plot(tokens, correspondance_df):
    
    tf_idf = TfidfVectorizer(tokenizer = tokenize)
    values = tf_idf.fit_transform(tokens.values())
    tsne = TSNE(n_components = 2, perplexity = 10, random_state = 0)
    tsne_result = tsne.fit_transform(values)
    
    tsne_df = pd.DataFrame({"tsne_1": tsne_result[:, 0],
                            "tsne_2": tsne_result[:, 1],
                            "label": correspondance_df["channel_title"],
                            "category": correspondance_df["channel_theme"]})
    
    plt.figure(figsize=(10, 10))
    plt.plot()
    sns.scatterplot(x = "tsne_1", y = "tsne_2", hue = "category", data = tsne_df)
    label_point(tsne_df["tsne_1"], tsne_df["tsne_2"], tsne_df["label"], plt.gca(), 3)  
    lim = (tsne_result.min() - 5, tsne_result.max() + 5)
    plt.gca().set_xlim(lim)
    plt.gca().set_ylim(lim)
    plt.gca().legend().remove()

    plt.gca().set_aspect("equal")
    plt.show()
```

```python
tokens = get_tokens(video_titles)
TSNE_plot(tokens, correspondance_dict_df)
```

```python
# PCA analysis

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pca_variables = ["channel_title", "channel_theme", "view_count", "video_count", "number_subscriber"]
pca_data = channels[pca_variables]

X = pca_data[["view_count", "video_count", "number_subscriber"]]
y = pca_data["channel_theme"]

X = StandardScaler().fit_transform(X)

pca = PCA(n_components = 2)
components = pca.fit_transform(X)

components_df = pd.DataFrame(data = components,
                             columns = ["PC1", "PC2"])
pca_df = pd.concat([components_df, y], axis = 1)

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1)
sns.scatterplot(x = pca_df["PC1"], y = pca_df["PC2"], hue = pca_df["channel_theme"])
label_point(pca_df["PC1"], pca_df["PC2"], pca_data["channel_title"], plt.gca(), 0.1)  

plt.show()

print(pca.explained_variance_ratio_)
```

```python
from sklearn.metrics.pairwise import linear_kernel

tf_idf = TfidfVectorizer(tokenizer = tokenize)
values = tf_idf.fit_transform(tokens.values())

cosine_similarities = linear_kernel(values, values)
print(cosine_similarities)
```

```python
# Channel recommandation based on videos title using cosine si

all_channels = correspondance_dict_df["channel_title"]

indices = pd.Series(all_channels.index, index=all_channels)

def channel_recommand(all_channels, channel_title):

    idx = indices[channel_title]

    sim_scores = list(enumerate(cosine_similarities[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:31]

    channel_indices = [i[0] for i in sim_scores]

    return all_channels.iloc[channel_indices]
```
